# Clean the Data

```python
# using .iloc to select certain columns
updated_apple_df = apple_df.iloc[:, [0,1,2,3,4,5]]
updated_apple_df.head()
```

```python
# using .iloc to select certain columns
updated_msft_df = microsoft_df.iloc[:, [0,1,2,3,4,5]]
updated_msft_df.head()
```

```python
# check for null values (count how many there are for apple dataset
updated_apple_df.isnull().sum()
```

```python
# check for null values (count how many there are for microsoft dataset
updated_msft_df.isnull().sum()
```

```python
# Merge the two DataFrames together based on the Dates they share
final_df = pd.merge(updated_apple_df, updated_msft_df, on="Date")
final_df.head()
```

```python
# Rename columns so that they are differentiated
final_df = final_df.rename(columns={"Open_x": "apple_open", "High_x": "apple_high", "Low_x": "apple_low",
                                      "Close(t)_x": "apple_close", "Volume_x": "apple_volume"})

final_df = final_df.rename(columns={"Open_y": "microsoft_open", "High_y": "microsoft_high", "Low_y": "microsoft_low",
                                      "Close(t)_y": "microsoft_close", "Volume_y": "microsoft_volume"})

final_df.head()
```

```python
# check datatypes
final_df.dtypes
```

```python
# change "Date" column to datetime using to_datetime() function
final_df['Date'] = pd.to_datetime(final_df['Date'])
final_df.dtypes
```

```python
# final preview of dataframe
final_df.head()
```

```python
# create class for "tech_stocks"

Base = declarative_base()

class techStocks(Base):
  __tablename__ = 'tech_stocks'
  date = Column(DateTime(), primary_key = True)
  apple_open = Column(Float())
  apple_high = Column(Float())
  apple_low = Column(Float())
  apple_close = Column(Float())
  apple_volume = Column(Integer())
  microsoft_open = Column(Float())
  microsoft_high = Column(Float())
  microsoft_low = Column(Float())
  microsoft_close = Column(Float())
  microsoft_volume = Column(Integer())
```
